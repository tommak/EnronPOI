### Feature selection

#### Univariate feature selection

```python
total_payment_const = ["salary", "bonus", "long_term_incentive",
                    "deferred_income", "deferral_payments", "loan_advances",
                     "other", "expenses" , "director_fees"]
total_stock_value_const = ["exercised_stock_options", "restricted_stock",
                    "restricted_stock_deferred"]

ordered_columns = total_payment_const + ["total_payments"] + \
            total_stock_value_const + ["total_stock_value"]

def clean(df):

    df.replace('NaN', np.nan, inplace=True)
    for c in (set(df.columns) - {"poi","email_address"}):
        df[c] = df[c].astype(np.float)

    df.drop("TOTAL", axis=0, inplace=True)
    df.drop("LOCKHART EUGENE E", axis=0, inplace=True)
    df.drop("THE TRAVEL AGENCY IN THE PARK", axis=0, inplace=True)


    fixed_df = df.copy()
    fix_pers = {
        "BELFER ROBERT" : 1,
        "BHATNAGAR SANJAY" : -1
    }

    ordered_columns = total_payment_const + ["total_payments"] + \
                total_stock_value_const + ["total_stock_value"]

    len_col = len(ordered_columns)
    for person, shift in fix_pers.items():
        for i in range(len_col):
            if i+shift<0 or i+shift>=len(ordered_columns):
                fixed_df.loc[person, ordered_columns[i]] = np.nan
            else:
                fixed_df.loc[person, ordered_columns[i]] = df.loc[person, ordered_columns[i+shift]]

    fixed_df.loc["BANNANTINE JAMES M", set(total_payment_const) - {"expenses"} ] = np.nan
    fixed_df.loc["BANNANTINE JAMES M", "total_payments"] = fixed_df.loc["BANNANTINE JAMES M", "expenses"]



    return fixed_df

def preprocess(dictionary):
    #Convert data frame to pandas data frame and perform cleaning
    df = pd.DataFrame.from_dict(dictionary, orient="index")
    fixed_df = clean(df)
    fixed_df["from_poi_perc"] = fixed_df["from_poi_to_this_person"]/fixed_df["to_messages"]
    fixed_df["to_poi_perc"] = fixed_df["from_this_person_to_poi"]/fixed_df["from_messages"]
    fixed_df["shared_with_poi_perc"] = fixed_df["shared_receipt_with_poi"]/fixed_df["to_messages"]
    
    fixed_df.replace(np.nan, "NaN", inplace=True)
    return fixed_df.to_dict(orient="index")

def preprocess_v4(dictionary):
    #Convert data frame to pandas data frame and perform cleaning
    df = pd.DataFrame.from_dict(dictionary, orient="index")
    fixed_df = clean(df)
    fixed_df["from_poi_perc"] = fixed_df["from_poi_to_this_person"]/fixed_df["to_messages"]
    fixed_df["to_poi_perc"] = fixed_df["from_this_person_to_poi"]/fixed_df["from_messages"]
    fixed_df["shared_with_poi_perc"] = fixed_df["shared_receipt_with_poi"]/fixed_df["to_messages"]
    
    
    fixed_df["gross_payments"] = fixed_df["total_payments"] - 2*fixed_df["deferred_income"]
    fixed_df["gross_stock_value"] =  fixed_df["total_stock_value"] - 2*fixed_df["restricted_stock_deferred"]

    for c in total_payment_const:
        fixed_df[c + "_perc"] = abs(fixed_df[c]) / fixed_df["gross_payments"]
    
    for c in total_stock_value_const:
        fixed_df[c + "_perc"] = abs(fixed_df[c]) / fixed_df["gross_stock_value"]

    fixed_df.replace(np.nan, "NaN", inplace=True)
    return fixed_df.to_dict(orient="index")



```

```python
features_list = ['poi'] + ordered_columns + ["to_messages", "from_messages",
                "from_poi_perc", "to_poi_perc", "shared_with_poi_perc"] #+ \
#                 [c + "_perc" for c in total_payment_const] + \
#                 [c + "_perc" for c in total_stock_value_const] + \
#                 ["gross_payments", "gross_stock_value"]



### Load the dictionary containing the dataset
with open("final_project_dataset.pkl", "r") as data_file:
    data_dict = pickle.load(data_file)

my_dataset = preprocess_v4(data_dict)

### Extract features and labels from dataset for local testing
data = featureFormat(my_dataset, features_list, sort_keys = True)
labels, features = targetFeatureSplit(data)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, \
                f1_score, make_scorer
```

```python
#clf = GaussianNB()


#clf = DecisionTreeClassifier(random_state=42, class_weight="balanced", min_samples_leaf=5)

base_clf = DecisionTreeClassifier(random_state=42, class_weight="balanced", min_samples_leaf=5)
param = {
    "class_weight" : ["balanced", None],
    "min_samples_leaf" : [3, 5, 8, 15],
    "criterion" : ["gini", "entropy"]
    
}


# base_est = DecisionTreeClassifier(random_state=42, class_weight="balanced")

# param = {
#     "base_estimator__max_depth" : [1, 2, 4, 6],
#     "n_estimators": [30, 50, 70, 100, 150, 200, 250]
# }

clf =  GridSearchCV(base_clf,
                    param_grid = param,
                    scoring = make_scorer(f1_score),
                   cv = StratifiedKFold(random_state=42))
```

```python
features_train, features_test, labels_train, labels_test = \
    train_test_split(features, labels, test_size=0.3, random_state=42, stratify=labels)
```

```python
print "Train dataset"
print sum([1. for val in labels_train if val == 1]) 
print sum([1. for val in labels_train if val == 1]) / len(labels_train)
print "Test dataset"
print sum([1. for val in labels_test if val == 1]) 
print sum([1. for val in labels_test if val == 1]) / len(labels_test)
```

```python
clf.fit(features_train, labels_train)
pred = clf.predict(features_test)

print "Accuracy: {}\n Precision: {}\n Recall: {}\n F1: {}"  \
                .format(accuracy_score(labels_test, pred),
                        precision_score(labels_test, pred),
                        recall_score(labels_test, pred),
                        f1_score(labels_test, pred))


```
